```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('results.csv')
```

```python
plt.hist(df['generations'], 100)
plt.xlabel("Generations until stable")
plt.ylabel("Count")
plt.title("Histogram of Generation Durations")
plt.show()
```

```python
metastable = df[(df['generations']<50000) & (df['generations']>1000)]
plt.hist(metastable['generations'], 100)
plt.xlabel("Generations until stable")
plt.ylabel("Count")
plt.title("Histogram of Generation Durations X-Cropped to (1000, 50000)")
plt.show()
```

```python
data = df
X = data['phappy']
Y = data['pneutral']
Z = data['punhappy']
generations = data['generations']

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(xs=X, ys=Y, zs=Z, cmap='bwr', c=generations)
ax.set_xlabel("p happy")
ax.set_ylabel("p neutral")
ax.set_zlabel("p unhappy")
fig.colorbar(None, ax=ax,cmap='bwr', label='Normalized Duration')
fig.set_figwidth(18)
plt.title("Sim Duration vs Birth Probabilities")
plt.show()
```

```python
survivors = df[df['generations']>50000]
fig, ax = plt.subplots(3, sharex=False)
fig.suptitle("Histograms of Birth Probs for Sims with Duration > 50k")
ax[0].hist(survivors['phappy'], 50)
ax[0].set_xlim((0.1,0.3))
ax[0].set_xlabel('p happy')
ax[1].hist(survivors['punhappy'], 50)
ax[1].set_xlim((0.1,0.3))
ax[1].set_xlabel('p unhappy')
ax[2].hist(survivors['pneutral'], 50)
ax[2].set_xlim((0.1,0.3))
ax[2].set_xlabel('p neutral')
fig.set_figheight(8)
for ax in fig.axes:
    ax.set_ylim((0, 20))
fig.subplots_adjust(hspace=0.3)
plt.show()
```

```python
thosewhodied = df[df['generations']<1000]
fig, ax = plt.subplots(3, sharex=False)
fig.suptitle("Histograms of Birth Probs for Sims with Duration < 1k")
ax[0].hist(thosewhodied['phappy'], 50)
ax[0].set_xlim((0.1,0.3))
ax[0].set_xlabel('p happy')
ax[1].hist(thosewhodied['punhappy'], 50)
ax[1].set_xlim((0.1,0.3))
ax[1].set_xlabel('p unhappy')
ax[2].hist(thosewhodied['pneutral'], 50)
ax[2].set_xlim((0.1,0.3))
ax[2].set_xlabel('p neutral')
for ax in fig.axes:
    ax.set_ylim((0, 20))
fig.set_figheight(8)
fig.subplots_adjust(hspace=0.3)
plt.show()
```
