```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('results.csv')
```

```python
plt.hist(df['generations'], 100)
plt.xlabel("Generations until stable")
plt.ylabel("Count")
plt.title("Histogram of Generation Durations")
plt.show()
```

```python

metastable = df[(df['generations']<50000) & (df['generations']>1000)]
plt.hist(metastable['generations'], 100)
plt.xlabel("Generations until stable")
plt.ylabel("Count")
plt.title("Histogram of Generation Durations X-Cropped to (1000, 50000)")

plt.show()
```

```python
data = df
X = data['phappy']
Y = data['pneutral']
Z = data['punhappy']
generations = data['generations']

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(xs=X, ys=Y, zs=Z, cmap='bwr', c=generations)
ax.set_xlabel("p happy")
ax.set_ylabel("p neutral")
ax.set_zlabel("p unhappy")
fig.colorbar(None, ax=ax,cmap='bwr', label='Normalized Duration')
fig.set_figwidth(18)
plt.title("Sim Duration vs Birth Probabilities")
plt.show()
```
